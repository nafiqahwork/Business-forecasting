# Background

## Business Context

You are working for an **online book platform** that offers both paid and free content. The company’s leadership wants to **differentiate** customers who are **content/experience-driven** (i.e., they buy books and engage deeply with them) from those who are **price-driven** (i.e., they focus on free content, sampling, or finding deals).

**Business Problem**: By identifying whether a customer is content- or price-driven, the marketing team can tailor promotions, upsell strategies, and new feature developments.

1. **Content/experience-driven customers** might respond better to:
   - Exclusive previews or early access to premium books.  
   - Advanced reading features (e.g., curated reading lists, advanced highlighting capabilities).

2. **Price-driven customers** might respond better to:
   - Discount offers or bundle deals.  
   - Extended free trials or “limited-time” free chapters.

---

## Why This Matters

- **Revenue Maximization**: If you misclassify a content-driven customer as price-driven, you might offer unneeded discounts and lose revenue.  
- **Customer Satisfaction**: Price-driven customers pushed with expensive premium items may feel alienated. Conversely, content-driven customers might be willing to pay more for better reading experiences—and you want to serve them accordingly.  
- **Long-term Loyalty**: Matching the right offers to the right user segment helps increase retention (reduce churn) and fosters loyalty.

---



# Dataset

## Data Files

- team_project_1.csv : This file contains data for past customers who used liveBook and whether or not they indicate in a survay that they care content more or price more. For each customer, there is a set of measurements about the use of liveBook by the customers. These measurements are described in more detail below.

## Variables in Data

The measurements of the customers, also known as the “features” of the dataset, are described below:

- Content Driven Customer: 1 if they indicate in a survay that they care content more, 0 if care price more.

- readingownedbook: This is how many events of reading a book the customer-owned.

- ebookdownloaded: This is how many times the customer downloads an e-book.

- readingfreepreview : This is how many times the customer read a free preview.

- highlightcreated: This is how many highlights the customer created.

- freecontentcheckout: This is how many times the customer used the free content check-out feature.

- readingopenchapter: This is how many times the customer read an open (free) chapter.

- wishlistitemadded: This is how many times the customer added items to their wish list.

- crossreferencetermopened: This is how many times the customer opened a cross-reference term.

- total_events: This is the total number of events logged for the customer.

- distinct_products: This is how many distinct products the customer purchased in the previous period.

- reading_utilities: This is how many times the customer used all reading utilities, including highlights and cross-reference terms.

- downloads_per_book: This is how many e-book downloads the customer had divided by the number of books they owned.

- percent_reading: This is the percent of the customers’ events that included reading.

- percent_utilities: This is the percent of the customers’ events that were utility features.

- events_per_book: This is the number of events the customer had divided by the distinct number of books they purchased.


## Data Collection Process

We recruited a select group of existing customers and offered them a modest incentive (e.g., gift cards, account credits) to complete a short survey about their primary motivation when purchasing books on our platform. The survey asked targeted questions such as:

* How important is reading quality, features (like highlighting or cross-referencing), and overall experience in your buying decision?
* How important are discounts, promotions, and overall price when making a purchase?
* Based on their answers, we categorized each respondent as either content/experience-driven or price-driven. This survey-labeled dataset now serves as a valuable “ground truth” for building and validating our predictive models on broader customer data.

# Task B : Regression with Interaction Terms

We use the first 20000 rows to fit regression model. We hold back the remaining rows for the subsequent model evaluation purpose.

```python
features = [
    'readingownedbook', 'ebookdownloaded', 'readingfreepreview',
    'highlightcreated', 'freecontentcheckout', 'readingopenchapter',
    'wishlistitemadded', 'crossreferencetermopened', 'total_events',
    'distinct_products', 'reading_utilities', 'downloads_per_book',
    'percent_reading', 'percent_utilities', 'events_per_book'
]
```

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/BF/team_project_1.csv')
df=df.iloc[:20000, :]
df.head()
```

```python
import pandas as pd
import itertools
import statsmodels.api as sm
import numpy as np

# Generate all possible pairs of features
feature_pairs = list(itertools.combinations(features, 2))

interaction_features = []
# Create interaction terms for each pair
for feature1, feature2 in feature_pairs:
    interaction_name = f"{feature1} x {feature2}"
    interaction_features.append(interaction_name)
    df[interaction_name] = df[feature1] * df[feature2]

# Display the first few rows of the DataFrame with the new interaction columns
df.head()
```

```python
print(interaction_features)
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

all_features = features + interaction_features

# Define X and y
X = df[all_features]
y = df['Content Driven Customer']

# Add a constant to the model (Adding y-intercept)
X = sm.add_constant(X)

# Fit the OLS model
model_interaction = sm.OLS(y, X).fit()

print(model_interaction.summary())

# Summary table as a DataFrame

summary_df = pd.DataFrame({
    "Variable": model.params.index,
    "Coef": model.params.values,
    "Std Err": model.bse.values,
    "t": model.tvalues,
    "P>|t|": model.pvalues,
    "CI 2.5%": model.conf_int()[0],
    "CI 97.5%": model.conf_int()[1]
})

summary_df = model_interaction.summary2().tables[1]
print(summary_df)
```

# **Q2. Which interactions turned out to be strong predictors?**

# A. Strong interaction predictors (p < 0.001 with high t-values) include:
- readingownedbook × ebookdownloaded: Coefficient = 0.0003, t ≈ 8.97
- readingownedbook × events_per_book: Coefficient = –6.481e-05, t ≈ –11.72
- ebookdownloaded × percent_reading: Coefficient = –0.0582, t ≈ –8.91
- downloads_per_book × percent_reading: Coefficient = 0.3288, t ≈ 15.89
- percent_reading × percent_utilities: Coefficient = 1.4523, t ≈ 11.92


These interactions indicate that the effect of one predictor significantly depends on another, enhancing the overall model’s explanatory power.

- readingownedbook × ebookdownloaded: Users who own physical books are also interested in eBooks. Rather than fully replacing them, they increased the engagement through digitalisation.

- readingownedbook × events_per_book: This shows an inverse relationship. Owning more books decreases the interest to engage in event because customers already have their preference on their own collection.

- ebookdownloaded × percent_reading: The negative coefficient indicates that ebooks tend to affect the reading percentage negatively. It does not result in increased reading activity.

- downloads_per_book × percent_reading: The strong positive relationship indicates that as number of download per book increases, the percent reading also increases. It indicates a high engagement in reading behavior among users.  

- percent_reading × percent_utilities: This shows that customers who spend more time on reading, use the reading utilities effictively. The high coefficient suggests that reading engagement relates with the digital tool usage and making this as the strongest key predictor in this model.
